```python
#I.
data_ciudades = {
    'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Habitantes': [1047996, 100169, 92331, 428450, 305717, 484475, 441649, 385098, 294395, 283734],
    'Area_km2': [2672.28, 1545.77, 1487.7, 116.5, 279.89, 6217.26, 368.9, 109.19, 59.4, 483.44],
    'Altitud_m': [154, 2325, 34, 3399, 29, 29, 106, 3271, 562, 156],
    'Densidad_poblacion_por_km2': [3924, 64.8, 62.1, 3673, 1091.6, 77.9, 1196, 3528.6, 4966, 587.2]
}

nfo = pd.DataFrame(data_ciudades)
nfo
```

```python
#II.
data_ciudades_2 = {
    'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Gentilicio': ['Limense', 'ArequipeÃ±o', 'Trujillano', 'CusqueÃ±o', 'Chiclayano', 'Piurano', 'IquiteÃ±o', 'Huancaino', 'TacneÃ±o', 'Pucallpino'],
    'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'],
    'Region': ['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'JunÃ­n', 'Tacna', 'Ucayali']
}
nombres = pd.DataFrame(data_ciudades_2)
nombres
```

```python
#III.
cuadro_1 = pd.merge(nfo, nombres, on = 'Ciudad', how = 'inner')
cuadro_1
```

```python
#IV. Mostramos las estadÃ­sticas descriptivas principales
cuadro_1.describe()
```

```python
#MÃ­nimas y MÃ¡xima Densidad
min_densidad = cuadro_1['Densidad_poblacion_por_km2'].min()
max_densidad = cuadro_1['Densidad_poblacion_por_km2'].max()
print(min_densidad)
print(max_densidad)
```

```python
#Ciudades respectivas
ciudad_min_densidad = cuadro_1[cuadro_1['Densidad_poblacion_por_km2'] == min_densidad]['Ciudad'].values[0]
ciudad_max_densidad = cuadro_1[cuadro_1['Densidad_poblacion_por_km2'] == max_densidad]['Ciudad'].values[0]
print(ciudad_min_densidad)
print(ciudad_max_densidad)
```

```python
#Statement
print(f"La mÃ­nima densidad poblacional la tiene {ciudad_min_densidad} con un valor de {min_densidad}.")
print(f"La mÃ¡xima densidad poblacional la tiene {ciudad_max_densidad} con un valor de {max_densidad}.")
```

```python
#V. Realizar un grÃ¡fico de barras donde se vea la cantidad de habitantes en cada ciudad
import matplotlib.pyplot as plt #importamos librerÃ­a
#Separamos los datos que queremos
habitantes = cuadro_1['Habitantes']
ciudades = cuadro_1['Ciudad']
```

```python
fig, ax = plt.subplots(figsize = (12, 6)) #Creamos la base para el grÃ¡fico
ax.plot()

ax.bar(x = ciudades, height = habitantes, color ='steelblue') #Ponemos los datos que irÃ¡n en los ejes

ax.set_xlabel('Ciudad') #Nombramos los ejes
ax.set_ylabel('Habitantes')
ax.set_title('NÃºmero de Habitantes por Ciudad de PerÃº') #Ponemos tÃ­tulo al grÃ¡fico
ax.tick_params(axis = "x", rotation = 90) #
plt.show()
```

```python
#VI. Realizar un grÃ¡fico de dispersiÃ³n entre la altura y el nÃºmero de habitantes en las ciudades del dataframe.
altitudes = cuadro_1['Altitud_m']

fig, ax = plt.subplots(figsize = (10,6))
ax.scatter(altitudes, habitantes, color = 'green')
ax.set_title('RelaciÃ³n entre Altitud y NÃºmero de Habitantes de las Ciudades en PerÃº')
ax.set_xlabel('Altitud (m)')
ax.set_ylabel('Habitantes')
ax.grid(True, alpha=0.5)
plt.show()
```
